```python
stop_word_count = []
word_count = []
sentence_count = []
unique_words = []

all_stopwords = stopwords.words('english')

for text in reviews['Text']:
    words = text.split(' ')
    stop_counter = 0
    sentences = ''
    
    for word in words:
        if word in all_stopwords:
            stop_counter += 1 
            
    word_count.append(len(words))
    stop_word_count.append(stop_counter)
    unique_words.append(len(set(words)))
    
    #number_of_sentenses
    sentences = sent_tokenize(text)
    sentence_count.append(len(sentences))
    
    
reviews['word_count'] = word_count
reviews['stopword_count'] = stop_word_count
reviews['sentence_count'] = sentence_count
reviews['unique_words'] = unique_words
```

```python
ps = PorterStemmer()
def clean_text(text) -> List:
    '''
    Given a raw text, this function cleans it by
        1. removing stop words, 
        2. punctuations that don't make sense by themselves (:,.!') 
            multiple punctuations without spaces are not removed.
        3. stems the words using porter stemmer 
     
    Input: raw text
    output: list of words in clean text
    '''
    words = text.split(' ')
    clean_text = list()
    
    for word in words:
        if word not in english_stop:
            if word not in string.punctuation:
                word = ps.stem(word)
                clean_text.append(word)
                
    return clean_text

def get_counts(text, label) -> Dict:
    '''
    Calculates how many times each word in the training dataset occurs in positive(1) and negative(0) classes
    
    Input: 
        text: Array of clean text from training data
        label: Corresponding labels from training data
    Output:
        Dictionary of frequency counts
    
    '''
    
    freqs = dict()
    
    for row in tqdm(zip(text, label)):
        text = row[0]
        label = row[1]

        for word in text:
            key = tuple((word,label))
            try:
                freqs[key] += 1
            except KeyError:
                freqs[key] = 1
                
    
    return freqs
```

```python
reviews['clean_text'] = reviews['Text'].progress_apply(lambda val: clean_text(val))
```

```python
freq_counts = get_counts(reviews['clean_text'], reviews['binary_label'])
```

```python
word_cnt_in_pos_data = list()
word_cnt_in_neg_data = list()

for i in range(len(reviews)):
    words_list = reviews['clean_text'][i]
    pos_cnt, neg_cnt = 0, 0
    for word in words_list:
            try:
                pos_cnt += freq_counts[tuple((word,0))]
            except KeyError:
                pos_cnt += 0

            try:
                neg_cnt += freq_counts[tuple((word,1))]
            except KeyError:
                neg_cnt += 0

    word_cnt_in_pos_data.append(pos_cnt)
    word_cnt_in_neg_data.append(neg_cnt)    


reviews['pos_cnt'] = word_cnt_in_pos_data
reviews['neg_cnt'] = word_cnt_in_neg_data 
```
